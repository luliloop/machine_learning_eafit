```python
#setting column names
col_names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','selection_probability']
df=pd.read_csv(filename,sep=',',header=None,names=col_names,parse_dates=True,index_col=None)
df.head()
```

```python
#dropping unnecesary columns
df.drop(['fnlwgt'], axis=1, inplace=True)
df.drop(['education-num'], axis=1, inplace=True)
df.drop(['occupation'], axis=1, inplace=True)
df.drop(['hours-per-week'], axis=1, inplace=True)
df.drop(['selection_probability'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.columns.str.strip()
```

```python
df.sex.replace({' Male': 0, ' Female': 1}, inplace=True)
df.head()
```

```python
workclass = pd.get_dummies(df.workclass)
workclass.head()
```

```python
df = pd.concat([df, workclass], axis=1)
df.head()
```
